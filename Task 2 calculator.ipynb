```python
print("This is a besic calculater")
x=int(input("enter your 1st number : "))
y=int(input("enter your 2nd number : "))
z=(input("enter your operator (+,-,*,/,**,//,%) : "))
if(z=="+"):
    print(x+y)
elif(z=="-"):
    print(x-y)
elif (z=='*'):
    print(x*y)
elif(z=="/" and y == 0):
    print(x,"can not divide by 0")
elif(z=="/"):
    print(x/y)
elif (z=="**"):
    print(x**y)
elif(z=="//"):
    print(x//y)
elif (z== '%'):
    print(x%y)
else:
    print("Enter valid operator")
```
