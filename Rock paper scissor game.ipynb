```python
import random

choices = ["rock", "paper", "scissors"]

win_rules = {
    "rock": "scissors",
    "scissors": "paper",
    "paper": "rock"
}

while True:
    user = input("Enter rock/paper/scissors or type 'exit' to quit: ").lower()

    if user == "exit":
        print("Game Ended ")
        break
    
    if user not in choices:
        print("Invalid choice!\n")
        continue
    
    computer = random.choice(choices)
    print("Computer chose:", computer)

    if user == computer:
        print("Draw!\n")
    elif win_rules[user] == computer:
        print("You Win! \n")
    else:
        print("Computer Wins!\n")
```
