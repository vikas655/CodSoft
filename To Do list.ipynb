```python
print("====================Welcome to my to do list=======================\n")
task=[]  #task list
while True:
    print("-----------Menu----------\n")
    print("1.Add Task")
    print("2.View Task")
    print("3.Delete Task")
    print("4.Exit")

    choise=int(input("enter your choise"))
    if(choise==1):
        date=(input("enter task date"))
        title=(input("enter task title"))
        discription=(input("enter task discription"))

        details={"date":date,"title":title,"discription":discription}

        task.append(details)
        print("Task added successfully")
    elif(choise==2):
        if(len(task)==0):
            print("no task added")

        else:
            print("view task\n")
            count=1  # giving numbers to task
            for i in task:
                print(count,":-->",i["date"], i['title'],
                      i["discription"])
                count+=1

    elif(choise==3):
        if len(task) == 0:
            print("No task to delete")
        else:
            print("Delete Task\n")
            count = 1
            for i in task:
                print(count, ":-->", i["title"])
                count += 1

            num = int(input("Enter task number to delete: "))

            if num >= 1 and num <= len(task):
                task.pop(num - 1) #indexing
                print("Task deleted successfully")
            else:
                print("Invalid task number")


    elif(choise==4):
        print("EXIT")
        break

    else:
        print("Enter valid task number")
```
